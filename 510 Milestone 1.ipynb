```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
responses = []

# Loop through three users.
for i in range(3):
    # Initialize a dictionary to store the responses for each user.
    response = {}
    response['Name'] = input('What is your name? ')
    response['Age'] = int(input('What is your age? '))
    response['Student'] = input('Are you a student? (yes/no) ')
    response['Sport'] = input('What is your favorite sport? ')
    responses.append(response)

df = pd.DataFrame(responses)

# Plot a bar chart showing the distribution of ages.
sns.countplot(x='Age', data=df)
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()

# Plot a bar chart showing the proportion of students and non-students.
sns.countplot(x='Student', data=df)
plt.title('Student Status')
plt.xlabel('Student')
plt.ylabel('Count')
plt.show()

# Plot a bar chart showing the favorite sports.
sns.countplot(x='Sport', data=df)
plt.title('Favorite Sports')
plt.xlabel('Sport')
plt.ylabel('Count')
plt.show()
```
